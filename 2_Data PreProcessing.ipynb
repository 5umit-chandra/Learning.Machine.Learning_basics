### Data set information 
**Number of Attributes(Column_Names)**: 11 (including the class)
```
1.  fLength:  continuous  - major axis of ellipse [mm]
2.  fWidth:   continuous  - minor axis of ellipse [mm] 
3.  fSize:    continuous  - 10-log of sum of content of all pixels [in #phot]
4.  fConc:    continuous  - ratio of sum of two highest pixels over fSize  [ratio]
5.  fConc1:   continuous  - ratio of highest pixel over fSize  [ratio]
6.  fAsym:    continuous  - distance from highest pixel to center, projected onto major axis [mm]
7.  fM3Long:  continuous  - 3rd root of third moment along major axis  [mm] 
8.  fM3Trans: continuous  - 3rd root of third moment along minor axis  [mm]
9.  fAlpha:   continuous  - angle of major axis with vector to origin [deg]
10.  fDist:   continuous  - distance from origin to center of ellipse [mm]
11.  class:   g,h         - gamma (signal), hadron (background)
```

## Encoding Categorical Data
Categorical data are variables that contain label values rather than numeric values. The number of possible values is often limited to a fixed set. 

Example values <u>"Yes" and "No"</u> cannot be used in mathematical equations. so we need to encode these variables into numbers. 
To achieve this we import `LabelEncoder` and `OneHotEncoder` class from `sklearn.preprocessing library`.

````python
# Replacing 'g' with '1' and 'h' with '0', and change the datatype to to integer
df["class"] = (df["class"] == "g").astype(int) # Boolean masks - efficient for large datasets

# df['class'] = df['class'].replace({'g': 0, 'h': 1}).astype(int) <-- Replace method
````

## Feature scaling / standardization

_Feature scaling ensures that features with varying magnitudes, units, and ranges do not bias machine learning algorithms. Most algorithms rely on Euclidean distance calculations, where features with higher magnitudes can disproportionately influence results._

**Feature standardization, such as Z-score normalization, transforms features to have zero mean and unit variance.** This equalizes the impact of each feature, enhancing algorithm performance. 

**The `StandardScaler from scikit-learn` is used for this purpose.**

**<u>StandardScaler normalizes the dataset's features by removing the mean and scaling each feature/variable to unit variance</u>**

````python
from sklearn.preprocessing import StandardScaler
````

    gamma: 7374, hadron: 4038
- train data contains 7374 instances of gamma and 4038 instances of hadron.
- Since the classes are imbalanced, with more gamma than hadron, **normalization is needed.**
- We need to oversample the hadron class by taking more instances from the gamma class.
- by using `from imblearn.over_sampling import RandomOverSampler`

````python
from imblearn.over_sampling import RandomOverSampler
````

    Preprocesses the dataset by scaling the features and optionally oversampling to handle class imbalance.
    
    Parameters:
    - dataframe: DataFrame: The input dataset.
    - oversample: bool, default=False: Flag indicating whether to perform oversampling.
    Returns:
    - data: array-like: The preprocessed dataset with scaled features and oversampled data if applicable.
    - X: array-like: Scaled features.
    - y: array-like: Target variable.

````python
# Extracting features (X) and target variable (y) from the DataFrame
def scale_dataset(dataframe, oversample=False):
  X = dataframe[dataframe.columns[:-1]].values
  y = dataframe[dataframe.columns[-1]].values

    # StandardScaling 
  scaler = StandardScaler()
  X = scaler.fit_transform(X)

    # normalizing gamma and hadron
  if oversample:
    ros = RandomOverSampler()
    X, y = ros.fit_resample(X, y)

     # Combining scaled features and target variable into a single array
  data = np.hstack((X, np.reshape(y, (len(y), 1))))

  return data, X, y
````

````python
# training data
train, X_train, y_train = scale_dataset(train, oversample=True)
# validation data
valid, X_valid, y_valid = scale_dataset(valid, oversample=False) # no oversampling
# test data
test, X_test, y_test = scale_dataset(test, oversample=False) # no oversampling
````

Oversampling is applied to the training data to address class imbalance.

Validation and test data are left unchanged to maintain the real-world distribution and prevent data leakage.

````python
# printing normalized instans in train dataset
print("y_train:", len(y_train))
print(f"gamma y_train: {sum(y_train ==1)}") #1 is gamma
print(f"hadron y_train: {sum(y_train ==0)}")
````

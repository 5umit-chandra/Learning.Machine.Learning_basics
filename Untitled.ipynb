# Dataset Information: [Seeds](https://archive.ics.uci.edu/dataset/236/seeds)
Folder: seeds/seeds_dataset.txt

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("seeds/seeds_dataset.txt", names=cols, sep="\s+") #separaters are spaces
cols = ["area", "perimeter", "compactness", "length", "width", "asymmetry", "groove", "class"]
```

```python
df
```

## EDA

```python
for i in range(len(cols)-1):
    for j in range(i+1, len(cols)-1):
        x_label = cols[i]
        y_label = cols[j]
        plt.scatter(df[x_label], df[y_label], c=df['class'], s=10)
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.show()
```

# Unsupervised Learning (PCA) and (K-mean)

### Dataset Information: [Seeds](https://archive.ics.uci.edu/dataset/236/seeds)
Folder: seeds/seeds_dataset.txt

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ["area", "perimeter", "compactness", "length", "width", "asymmetry", "groove", "class"]
df = pd.read_csv("seeds/seeds_dataset.txt", names=cols, sep="\s+") #separaters are spaces
```

```python
df
```

```python
df.isnull().sum()
```

## Exploratory Data Analysis (EDA) 

```python
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning) # Suppress FutureWarnings

for i in range(len(cols)-1):
    for j in range(i+1, len(cols)-1):
        x_label = cols[i]
        y_label = cols[j]
        sns.scatterplot(x=x_label, y=y_label, data=df, hue='class')
        plt.show()
```

### Similarly to the scatter plots above but created using matplotlib.

```python
for i in range(len(cols)-1):
    for j in range(i+1, len(cols)-1):
        plt.figure(figsize=(5, 3))
        x_label = cols[i]
        y_label = cols[j]
        
        # Get unique categories of 'class'
        unique_classes = df['class'].unique()
        
        # Plot each category separately with a different color
        for class_label in unique_classes:
            class_data = df[df['class'] == class_label]
            plt.scatter(class_data[x_label], class_data[y_label], s=10, alpha=0.6, label=f'class {class_label}')
        
        plt.xlabel(x_label)
        plt.ylabel(y_label)
        plt.legend()  # legend
        plt.show()
```

# K-means clustering

<img src="img/k-mean_cluttering.png" width="600" height="300">
<img src="img/k-mean_cluttering_2.png" width="600" height="300">
groups data into clusters to minimize the sum of squared distances between objects and their assigned cluster means

```python
from sklearn.cluster import KMeans
```

```python
# data preparation
x = "perimeter"
y = "asymmetry"
X = df[[x, y]].values
```

```python
import os
```

```python
os.environ["OMP_NUM_THREADS"] = '1'  # to avoid memory leak when dealing with KMeans
```

```python
warnings.filterwarnings("ignore", category=UserWarning) # Suppress the warning
```

```python
kmeans = KMeans(n_clusters=3, random_state=0, n_init='auto').fit(X)
```

```python
clusters = kmeans.labels_
```

```python
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1, 1))), columns=[x, y, "class"])
```

```python
# K Means classes
sns.scatterplot(x=x, y=y, hue='class', data=cluster_df)
plt.plot()
```

```python
# Original classes
sns.scatterplot(x=x, y=y, hue='class', data=df)
plt.plot()
```

```python
clusters
```

```python
df['class'].values
```

```python
# subplots side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# K Means classes
sns.scatterplot(x=x, y=y, hue='class', data=cluster_df, ax=axs[0])
axs[0].set_title('K Means classes')

# Original classes
sns.scatterplot(x=x, y=y, hue='class', data=df, ax=axs[1])
axs[1].set_title('Original classes')


plt.show()
```

0.0 in k-means class represents 1 in original, 1.0 is 2 , and 2.0 is 3 .... we can map them if necessary

## Higher Dimensions

```python
X = df[cols[:-1]].values
```

```python
kmeans = KMeans(n_clusters = 3).fit(X)
cluster_df = pd.DataFrame(np.hstack((X, kmeans.labels_.reshape(-1, 1))), columns=df.columns)
```

```python
# K Means classes
sns.scatterplot(x=x, y=y, hue='class', data=cluster_df)
plt.plot()
```

```python
# Original classes
sns.scatterplot(x=x, y=y, hue='class', data=df)
plt.plot()
```

# Principal Component Analysis (PCA)

<img src="img/Principal_component_analysis(PCA).png" width="700" height="350">
(PCA) is a dimensionality reduction technique

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
transformed_x = pca.fit_transform(X)
```

```python
X.shape
```

```python
transformed_x.shape
```

```python
transformed_x[:5]
```

```python
plt.scatter(transformed_x[:,0], transformed_x[:,1])
plt.show()
```

```python
kmeans_pca_df = pd.DataFrame(np.hstack((transformed_x, kmeans.labels_.reshape(-1, 1))), columns=["pca1", "pca2", "class"])
```

```python
truth_pca_df = pd.DataFrame(np.hstack((transformed_x, df["class"].values.reshape(-1, 1))), columns=["pca1", "pca2", "class"])
```

```python
# K Means classes
sns.scatterplot(x="pca1", y="pca2", hue='class', data=kmeans_pca_df)
plt.plot()
```

```python
# Truth classes
sns.scatterplot(x="pca1", y="pca2", hue='class', data=truth_pca_df)
plt.plot()
```

```python
# color palette
color_palette = sns.color_palette('hls', 3)    #or(['#1f77b4', '#ff7f0e', '#2ca02c'])

# figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# K Means classes
sns.scatterplot(x="pca1", y="pca2", hue='class', data=kmeans_pca_df, palette=color_palette, ax=axs[0])
axs[0].set_title('K Means classes')

# Truth classes
sns.scatterplot(x="pca1", y="pca2", hue='class', data=truth_pca_df, palette=color_palette, ax=axs[1])
axs[1].set_title('Truth classes')


plt.show()
```
